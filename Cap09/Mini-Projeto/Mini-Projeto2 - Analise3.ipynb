```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
df.head(5).T
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
```

```python
plt.figure(figsize=(12, 8), dpi=200)

fontsize = 14
titlesize = 24

order_gearbox = df.groupby('gearbox').count().index
order_typeveh = df.groupby('fuelType').count().index

fig_sns = sns.barplot(x="fuelType", y="price", hue="gearbox", order = order_typeveh, hue_order= order_gearbox, \
                      data=df, palette="Paired", estimator= np.mean )
plt.xlabel('Tipo de Combustível', fontsize=fontsize)
plt.ylabel('Preço Médio', fontsize=fontsize)
plt.title('Preço médio dos veículos por tipo de combustível e tipo de caixa de câmbio', fontsize=titlesize, color='r')


media = df.groupby(['gearbox','fuelType'])['price'].mean().astype(int)
for p, label in zip(fig_sns.patches, media):
    fig_sns.annotate(label, (p.get_x()+0, p.get_height()+300))

plt.show()
```

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(8,5))
colors = ["#00e600", "#ff8c1a","#a180cc"]
sns.barplot(x="fuelType", y="price",hue="gearbox", palette="husl",data=df)
ax.set_title("Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio",fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de Combustível",fontdict= {'size':14})
ax.yaxis.set_label_text("Preço Médio",fontdict= {'size':14})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
fig = fig_sns.get_figure()
fig.savefig("plots/Analise3/fueltype-vehicleType-price-my.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
```

```python
plt.figure(figsize=(12, 8), dpi=200)

fontsize = 14
titlesize = 24

order_gearbox = df.groupby('gearbox').count().index
order_typeveh = df.groupby('vehicleType').count().index

fig_sns = sns.barplot(x="vehicleType", y="powerPS", hue="gearbox", order = order_typeveh, hue_order= order_gearbox, \
                      data=df, palette="colorblind", estimator= np.mean )
plt.xlabel('Tipo de Veículos', fontsize=fontsize)
plt.ylabel('Preço Médio', fontsize=fontsize)
plt.title('Preço médio dos veículos por tipo de veículo e tipo de caixa de câmbio', fontsize=titlesize, color='r')


media = df.groupby(['gearbox','vehicleType'])['powerPS'].mean().astype(int)
for p, label in zip(fig_sns.patches, media):
    fig_sns.annotate(label, (p.get_x()+0, p.get_height()+5))

plt.show()
```

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
colors = ["windows blue", "amber", "greyish", "faded green", "dusty purple"]
fig, ax = plt.subplots(figsize=(8,5))
sns.set_palette(sns.xkcd_palette(colors))
sns.barplot(x="vehicleType", y="powerPS",hue="gearbox",data=df)
ax.set_title("Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio",fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':14})
ax.yaxis.set_label_text("Potência Média",fontdict= {'size':14})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
fig = fig_sns.get_figure()
fig.savefig("plots/Analise3/vehicletype-fueltype-power-my.png")
```
