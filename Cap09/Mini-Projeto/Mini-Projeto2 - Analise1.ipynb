## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
#%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path, encoding="latin-1")
df.head(5).T
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
```

```python
plt.figure(figsize=(8, 6), dpi=500)

fontsize = 15
fig_sns = sns.distplot(df['yearOfRegistration'], color='#33cc33')
plt.xlabel('Ano de Registro', fontsize=fontsize)
plt.ylabel('Densidade (KDE)', fontsize=fontsize)
plt.title('Distribuição de Veículos com base no Ano de Registro', fontsize=fontsize)
plt.show()
```

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(df["yearOfRegistration"], color="#33cc33",kde=True, ax=ax)
ax.set_title('Distribuição de Veículos com base no Ano de Registro', fontsize= 15)
plt.ylabel("Densidade (KDE)", fontsize= 15)
plt.xlabel("Ano de Registro", fontsize= 15)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution.png")
fig = fig_sns.get_figure()
fig.savefig("plots/Analise1/vehicle-distribution-my.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
#Crie um Boxplot para avaliar os outliers
plt.figure(figsize=(12, 8), dpi=200)
sns.set_style("whitegrid")

fontsize = 14
titlesize = 24
fig_sns = sns.boxplot(x = "vehicleType", y = "price", data = df)
plt.text(5.25,27000,"Análise de Outliers",fontsize=18,color="r",ha="center", va="center")

plt.xlabel('Tipo de Veículo', fontsize = 14)
plt.ylabel('Range de Preço', fontsize = fontsize)
#plt.title('Análise de Outlier', fontsize=titlesize, color='r')
plt.show()
```

```python
# Crie um Boxplot para avaliar os outliers
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(x="vehicleType", y="price", data=df)
ax.text(5.25,27000,"Análise de Outliers",fontsize=18,color="r",ha="center", va="center")
ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':14})
ax.yaxis.set_label_text("Range de Preço",fontdict= {'size':14})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png")
fig = fig_sns.get_figure()
fig.savefig("plots/Analise1/price-vehicleType-boxplot-my.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
plt.figure(figsize=(12, 8), dpi=200)
sns.set_style("white")

fontsize = 14
titlesize = 24
fig_sns = sns.countplot(x = "vehicleType", data = df, palette="BuPu", order = df['vehicleType'].value_counts().index)
plt.xlabel('Tipo de Veículo', fontsize=fontsize)
plt.ylabel('Total de Veículos para Vendas', fontsize=fontsize)
plt.title('Contagem total de veículos à venda conforme o tipo de veículo', fontsize=titlesize, color='r')


for p, label in zip(fig_sns.patches, df["vehicleType"].value_counts()):
    fig_sns.annotate(label, (p.get_x()+0.25, p.get_height()+0.15))
    
plt.show()
```

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
sns.set_style("white")
g = sns.catplot(x="vehicleType", data=df, kind="count", palette="BuPu", height=6, aspect=1.5)
g.ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Total de Veículos Para Venda", fontdict= {'size':16})
g.ax.set_title("Contagem total de veículos à venda conforme o tipo de veículo",fontdict= {'size':18})
# to get the counts on the top heads of the bar
for p in g.ax.patches:
    g.ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+500))
```

```python
# Salvando o plot
g.savefig("plots/Analise1/count-vehicleType-my.png")
fig = fig_sns.get_figure()
fig.savefig("plots/Analise1/count-vehicleType-my.png")
```
